# Data Cleaning

In this section, you will need to read the CSV files into DataFrames and perform any necessary data cleaning steps. After cleaning, combine all DataFrames into a single DataFrame.

Files:

* `whale_returns.csv`: Contains returns of some famous "whale" investors' portfolios.

* `algo_returns.csv`: Contains returns from the in-house trading algorithms from Harold's company.

* `sp_tsx_history.csv`: Contains historical closing prices of the S&P TSX 60 Index.

## Whale Returns

Read the Whale Portfolio daily returns and clean the data.

```python
# Reading whale returns
whale_returns_data = Path("./Resources/whale_returns.csv")
whale_returns = pd.read_csv(whale_returns_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
```

```python
# Count nulls
whale_returns.isnull().sum()
```

```python
# Drop nulls
whale_returns = whale_returns.dropna().copy()
whale_returns.head()
```

## Algorithmic Daily Returns

Read the algorithmic daily returns and clean the data.

```python
# Reading algorithmic returns
algo_returns_data = Path("./Resources/algo_returns.csv")
algo_returns = pd.read_csv(algo_returns_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
algo_returns.head()
```

```python
# Count nulls
algo_returns.isnull().sum()
```

```python
# Drop nulls
algo_returns = algo_returns.dropna().copy()
algo_returns.head()
```

## S&P TSX 60 Returns

Read the S&P TSX 60 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P TSX 60 Closing Prices
sp_tsx_history_data = Path("./Resources/sp_tsx_history.csv")
sp_tsx_history = pd.read_csv(sp_tsx_history_data, index_col="Date", infer_datetime_format=True, parse_dates=True)
sp_tsx_history.head()
```

```python
# Check Data Types
sp_tsx_history.dtypes
```

```python
# Remove "$" and ","
sp_tsx_history["Close"] = sp_tsx_history["Close"].str.replace("$", "", regex=True)
sp_tsx_history["Close"] = sp_tsx_history["Close"].str.replace(",", "", regex=True)
# Fix Data Types
sp_tsx_history["Close"] = sp_tsx_history["Close"].astype("float")
sp_tsx_history.head()
```

```python
# Calculate Daily Returns
sp_tsx_history_returns = sthr = sp_tsx_history.pct_change()
sthr.head()
```

```python
# Drop nulls
sthr = sthr.dropna().copy()
sthr.head()
```

```python
# Rename `Close` Column to be specific to this portfolio.
sthr = sthr.rename(columns={"Close": "S&P TSX 60 Returns"})
sthr.head()
```

## Combine Whale, Algorithmic, and S&P TSX 60 Returns

```python
# Join Whale Returns, Algorithmic Returns, and the S&P TSX 60 Returns into a single DataFrame with columns for each portfolio's returns.
all_returns = pd.concat([whale_returns, algo_returns, sthr], axis="columns", join="inner")
all_returns.head()
```

### Determine which portfolios are riskier than the S&P TSX 60

```python
# Calculate the daily standard deviation of S&P TSX 60
sp_tsx_60_std_raw = cumulative_returns[['S&P TSX 60 Returns']].std()
sp_tsx_60_std = float(sp_tsx_60_std_raw)
print(f'The daily standard deviation of S&P TSX 60 is {sp_tsx_60_std:.4f}')
# Determine which portfolios are riskier than the S&P TSX 60
soros_fund_management_raw = cumulative_returns[['SOROS FUND MANAGEMENT LLC']].std()

soros_fund_management_std = float(soros_fund_management_raw)

paulson_co_raw = cumulative_returns[['Paulson&co']].std()

paulson_co_std = float(paulson_co_raw)

tiger_global_raw = cumulative_returns[['TIGER GLOBAL MANAGEMENT LLC']].std()

tiger_global_std = float(tiger_global_raw)

berkshire_hathaway_raw = cumulative_returns[['BERKSHIRE HATHAWAY INC']].std()

bershire_hathaway_std = float(berkshire_hathaway_raw)

algo_1_raw = cumulative_returns[['Algo 1']].std()

algo_1_std = float(algo_1_raw)

algo_2_raw = cumulative_returns[['Algo 2']].std()

algo_2_std = float(algo_2_raw)

if soros_fund_management_std > sp_tsx_60_std:
    print('SOROS FUND MANAGEMENT LLC is riskier than S&P TSX 60')
elif soros_fund_management_std < sp_tsx_60_std:
    print('SOROS FUND MANAGEMENT LLC is less riskier than S&P TSX 60')
if paulson_co_std > sp_tsx_60_std:
    print('PAULSON & CO.INC. is riskier than S&P TSX 60')
elif paulson_co_std < sp_tsx_60_std:
    print('PAULSON & CO.INC. is less riskier than S&P TSX 60')
if tiger_global_std > sp_tsx_60_std:
    print('TIGER GLOBAL MANAGEMENT LLC is riskier than S&P TSX 60')
elif tiger_global_std < sp_tsx_60_std:
    print('TIGER GLOBAL MANAGEMENT LLC is less riskier than S&P TSX 60')
if bershire_hathaway_std > sp_tsx_60_std:
    print('BERKSHIRE HATHAWAY INC is riskier than S&P TSX 60')
elif bershire_hathaway_std < sp_tsx_60_std:
    print('BERKSHIRE HATHAWAY INC is less riskier than S&P TSX 60')
if algo_1_std > sp_tsx_60_std:
    print('Algo 1 is riskier than S&P TSX 60')
elif algo_1_std < sp_tsx_60_std:
    print('Algo 1 is less riskier than S&P TSX 60')
if algo_2_std > sp_tsx_60_std:
    print('Algo 2 is riskier than S&P TSX 60')
elif algo_2_std < sp_tsx_60_std:
    print('Algo 1 is less riskier than S&P TSX 60')
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
annualized_std = cumulative_std * np.sqrt(252)
annualized_std
```

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
all_returns_rolling_std = all_returns.rolling(window=21).std()
# Plot the rolling standard deviation
all_returns_rolling_std.plot()
```

### Calculate and plot the correlation

```python
# Calculate the correlation
correlation = all_returns.corr()
# Display de correlation matrix
correlation.head()
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 60 TSX

```python
# Calculate covariance of a single portfolio
covariance = all_returns['Paulson&co'].cov(all_returns['S&P TSX 60 Returns'])
print(f'Covariance of PAULSON & CO.INC. is {covariance}')
# Calculate variance of S&P TSX
sp_tsx_var = all_returns['S&P TSX 60 Returns'].var()
print(f'Variance of S&P TSX is {sp_tsx_var}')
# Computing beta
paulson_co_beta = covariance / sp_tsx_var
print(paulson_co_beta)
# Plot beta trend
rolling_covariance = all_returns['Paulson&co'].rolling(window=30).cov(all_returns['S&P TSX 60 Returns'])
rolling_covariance.plot(figsize=(20, 10), title='Rolling 30-Day Covariance of PAULSON & CO.INC. Returns vs. S&P TSX 60 Returns')
```

This is like a moving window average, but it assigns greater importance to more recent observations. Try calculating the [`ewm`](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.ewm.html) with a 21-day half-life.

```python
# Use `ewm` to calculate the rolling window
exponentially_weighted_average_21_days = ewa = all_returns['Paulson&co'].ewm(span=4, adjust=False).mean()
ewa.plot()
```

# Sharpe Ratios
In reality, investment managers and thier institutional investors look at the ratio of return-to-risk, and not just returns alone. After all, if you could invest in one of two portfolios, and each offered the same 10% return, yet one offered lower risk, you'd take that one, right?

### Using the daily returns, calculate and visualize the Sharpe ratios using a bar plot

```python
# Annualized Sharpe Ratios
sharpe_ratios = (all_returns.mean() * 252) / (cumulative_std * np.sqrt(252))
print(sharpe_ratios)
```

```python
# Visualize the sharpe ratios as a bar plot
sharpe_ratio_dict = dict(sharpe_ratios)
stock = list(sharpe_ratio_dict.keys())
sharpe_ratios = list(sharpe_ratio_dict.values())
plt.bar(stock, sharpe_ratios, color = 'blue', width = 0.4)
plt.xticks(rotation = 90)
```
